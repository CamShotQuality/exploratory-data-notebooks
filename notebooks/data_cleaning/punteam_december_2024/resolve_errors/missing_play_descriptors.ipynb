```python
# save game id/play id combinations to a variable
game_play_ids = december_2024_nba_pbp_df[
        (december_2024_nba_pbp_df['play_descriptors'] == '[]') &
        (december_2024_nba_pbp_df['action_type'].isin(['2pt', '3pt']))
    ][['game_id', 'play_id']]

december_2024_nba_pbp_df[
        (december_2024_nba_pbp_df['play_descriptors'] == '[]') &
        (december_2024_nba_pbp_df['action_type'].isin(['2pt', '3pt']))
    ]
```

```python
# use the game id/play id combination to query for play_descriptors 
# 
# use game_play_ids as parameters 

plays_missing_descriptors = pd.read_sql(
    sa.text(
        """
        SELECT public_clean_descriptors
        FROM shotquality.play_by_play_t
        WHERE game_id = :game_id AND play_id = :play_id
        """
    ),
    load_connection_string("sqdb"),
    params={
        "game_ids": tuple(game_play_ids['game_id'].unique().tolist()),
        "play_ids": tuple(game_play_ids['play_id'].unique().tolist())
    }
)
```

```python
# join result to december_2024_nba_pbp_df and add to play_descriptors column
december_2024_nba_pbp_df = december_2024_nba_pbp_df.merge(plays_missing_descriptors, on=['game_id', 'play_id'], how='left')

# save to a CSV
# december_2024_nba_pbp_df.to_csv('december_2024_nba_pbp_with_descriptors.csv', index=False)

december_2024_nba_pbp_df.head()
```
